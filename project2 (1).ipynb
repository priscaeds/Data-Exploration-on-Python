Data Exploration in python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df= pd.read_csv('product.csv')
df
```

1.create 3 calculated columns

```python
# i.Revenue

df['Revenue'] = df['Units_Sold'] * df['Price']
df
```

```python
# ii. Total_cost

df['Total_cost'] = df['Units_Sold'] * df['Cost']
df
```

```python
# iii. profit

df['Profit']= (df['Revenue'] - df['Total_cost']) 
df
```

2.Perform correlation analysis using a heatmap

```python
# select columns for the analysis

xyz = ['Units_Sold', 'Price','Profit']
corr = df[xyz].corr()
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(corr, annot= True, cmap='coolwarm', square=True)
plt.title('correlation Heatmap')
plt.show()
```

7.Create a linear, and logistic regression model to make predictions on a dataset.

```python
# linear regression
# import the libraries

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# define features x and y

x=df[['Price']]
y=df['Profit']
```

```python
# split data into train and test set

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
# create regression model

model= LinearRegression()
```

```python
# train the model

model.fit(x_train, y_train)
```

```python
# make prediction

y_pred =model.predict(x_test)
```

```python
# logistic regression
# import the library

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
#Add a new column to indicate whether a product was purchased based on its revenue

df['Purchased'] = df['Revenue'].apply(lambda x: 1 if x > 3000 else 0)
```

```python
df
```

```python
# split into test and train also

x_train, x_test, y_train, y_test = train_test_split(df[['Revenue']], df.Purchased, test_size=0.2, random_state=42)
x_test
```

```python
# create and train the model

model= LogisticRegression()
model.fit(x_train, y_train)
```

```python
# make a prediction on the x_test

model.predict(x_test)
```

```python
# check the score (model accuracy), if the output is 1 then the model is perfect.

model.score(x_test,y_test)
```

```python
# predict if there's a purchase at the revenue of 14700

model.predict(np.array([[14700.00]]))
```
